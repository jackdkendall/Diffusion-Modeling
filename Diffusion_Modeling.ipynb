```python
from __future__ import division
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
%matplotlib inline
```

```python
D = 1.2e-15
L = 1/10000
M_vec = []
times = range(100000)
```

```python
for a in range(5,50,5):
    M = []
    for t in times:
        S = np.zeros(a)
        for n in range(a):
            S[n] = (1/((2*(n)+1)**2))*np.exp(((-D)*((2*(n)+1)**2)*(np.pi**2)*(t))/L**2)
        M.append(1-((8/np.pi**2)*np.sum(S)))
    M_vec.append(M)
```

```python
plt.plot(times[0:10000], np.transpose(M_vec)[0:10000])
plt.show()
```

```python
def Diffusion(times, D):
    a = 30
    L = 1/10000
    M = []
    
    for t in times:
        S = np.zeros(a)
        for n in range(a):
            S[n] = (1/((2*(n)+1)**2))*np.exp(((-D)*((2*(n)+1)**2)*(np.pi**2)*(t))/L**2)
        M.append((1-((8/np.pi**2)*np.sum(S))))
    return M
```

```python
x = np.array([2,4,8,24,48,72,96,120,144,168])
x *= 3600
y = np.array([0.081102667, 0.596991, 4.666506, 14.83437333, 23.223108, 29.575486, 32.873329, 33.08912267, 33.24061033, 33.49927467])
y /= 100
```

```python
plt.plot(x,y)
plt.show()
```

```python
D_opt = curve_fit(Diffusion, x, y, p0=1.2e-15)
```
